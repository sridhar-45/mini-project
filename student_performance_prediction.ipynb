##  Forward Fill or Backward Fill
Forward Fill (ffill): Fills NaN values with the previous non-null value.

Backward Fill (bfill): Fills NaN values with the next non-null value.

```python
# Forward fill: Fill NaN with the previous value
data_filled_ffill = data.fillna(method='ffill')

# Backward fill: Fill NaN with the next value
data_filled_bfill = data.fillna(method='bfill')

# View the filled data
print(data_filled_ffill.head())
print(data_filled_bfill.head())
```

## Filling Null Values with Mean, Median, or Mode:
For numerical columns, it's common to fill missing values with the mean or median. For categorical columns, you can fill with the mode (most frequent value).

## Filling with Mean (for numerical columns):

```python
# Fill NaN values in numerical columns with the mean of each column
data_filled_mean = data.fillna(data.mean())
print(data_filled_mean.head())

# Fill NaN values in numerical columns with the median of each column
data_filled_median = data.fillna(data.median())
print(data_filled_median.head())

# Fill NaN values in categorical columns with the mode (most frequent value)
# data.mode().iloc[0]: Fills missing values in categorical columns with the most frequent value.
data_filled_mode = data.fillna(data.mode().iloc[0])
print(data_filled_mode.head())

```

#2. Splitting the Dataset into Features (X) and Target (y)
 Separate the dataset into independent variables (features) and the dependent variable (target).

 1. data.drop(columns=['G3']): Removes the G3 column from the dataset, leaving all other columns as the feature set X.
 2. data['G3']: Selects the G3 column as the target variable y (e.g., the grade we're predicting).

```python
x = data.drop(columns = ['G3'])
y = data['G3']
```

# 3. Train-Test Split
Divide the data into a training set (to train the model) and a testing set (to evaluate the model).

**1. train_test_split:** Splits the dataset into training and testing subsets.

**2. test_size=0.2:** Allocates 20% of the data to the testing set and 80% to the training set.
**3. random_state=42:** Ensures reproducibility by setting a seed for the random number generator.

### outputs

X_train: Features for training (80% of X).

X_test: Features for testing (20% of X).

y_train: Target values for training (80% of y).

y_test: Target values for testing (20% of y).



```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.2, random_state = 42 )
```

# 4. Model Training and Evaluation


## Linear regression :     
Linear Regression is a simple and interpretable machine learning model that assumes a linear relationship between the features (X) and the target variable (y).

How it works:

* Fits a straight line (y = mx + b) that minimizes the sum of squared differences
between actual and predicted values.

* The model calculates the weights (m) and intercept (b) for the line of best fit.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

model = LinearRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f"Linear Regression :  MSE = {mse : .2f} , R2 = {r2 : .2f}")
```

# 2. Random Forest Regressor :     
Random Forest is an ensemble learning method that combines multiple Decision Trees to improve prediction accuracy.

* Builds many Decision Trees using random subsets of data and features.
* Each tree makes a prediction, and the Random Forest averages these predictions (regression) to produce the final output.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error,r2_score

model = RandomForestRegressor(random_state = 42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f"Random Forest: MSE={mse:.2f}, RÂ²={r2:.2f}")
```

# 3. Decision Tree Regressor
A Decision Tree is a tree-like structure where each node splits the data based on a feature that reduces the prediction error the most.

* Recursively splits the dataset into smaller subsets using conditions on feature values.
* The final prediction is the average of target values in a leaf node.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,r2_score

model = DecisionTreeRegressor(random_state = 42)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f"Decision Tree: MSE={mse:.2f}, RÂ²={r2:.2f}")
```

# 4. Support Vector Regressor (SVR)
Support Vector Regression (SVR) uses Support Vector Machines (SVMs) for regression tasks. It aims to fit a model within a margin of tolerance (epsilon) around the target values.

**How it works:**

* Maps data into a high-dimensional space using kernels (linear, polynomial, RBF, etc.).
* Finds the hyperplane that minimizes error while staying within a tolerance margin (epsilon).

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,r2_score

model = SVR()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print(f"Support Vector Regressor: MSE={mse:.2f}, RÂ²={r2:.2f}")
```

# Train and Evaluate Classification Models (for predicting pass/fail):

## 1. Binary Classification
Convert the target variable y (grades) into a binary classification problem.
If the grade (y) is greater than or equal to 10, classify it as 1 (Pass).
Otherwise, classify it as 0 (Fail).

**Why Use Binary Classification?:**

Binary classification is often used for simpler, pass/fail predictions instead of predicting exact numerical grades.

```python
import numpy as np

y_binary = np.where(y >= 10, 1, 0)
y_train_binary = np.where(y_train >= 10, 1, 0)
y_test_binary = np.where(y_test >= 10, 1, 0)
```

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(x_train, y_train_binary)  # Train the model
y_pred_lr = lr_model.predict(x_test)  # Predict on test data

# Evaluation
lr_accuracy = accuracy_score(y_test_binary, y_pred_lr)
print(f"Logistic Regression - Accuracy: {lr_accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test_binary, y_pred_lr))
```

# 2. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf_model = RandomForestClassifier(random_state = 42)
rf_model.fit(x_train, y_train_binary)
y_pred_rf = rf_model.predict(x_test)

rf_accuracy = accuracy_score(y_test_binary, y_pred_rf)
print(f"Random Forest - Accuracy: {rf_accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test_binary, y_pred_rf))
```

# 3. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(x_train, y_train_binary)  # Train the model
y_pred_dt = dt_model.predict(x_test)  # Predict on test data

# Evaluation
dt_accuracy = accuracy_score(y_test_binary, y_pred_dt)
print(f"Decision Tree - Accuracy: {dt_accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test_binary, y_pred_dt))
```

# 4. Support Vector Classifier (SVC)

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Support Vector Classifier
svc_model = SVC()
svc_model.fit(x_train, y_train_binary)  # Train the model
y_pred_svc = svc_model.predict(x_test)  # Predict on test data

# Evaluation
svc_accuracy = accuracy_score(y_test_binary, y_pred_svc)
print(f"Support Vector Classifier - Accuracy: {svc_accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test_binary, y_pred_svc))
```

## classification models

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

# Convert grades to pass/fail (binary classification)
y_binary = np.where(y >= 10, 1, 0)
y_train_binary = np.where(y_train >= 10, 1, 0)
y_test_binary = np.where(y_test >= 10, 1, 0)

# Initialize models
classifiers = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Support Vector Classifier": SVC()
}

# Train and evaluate each model
for name, model in classifiers.items():
    model.fit(x_train, y_train_binary)
    y_pred_binary = model.predict(x_test)
    accuracy = accuracy_score(y_test_binary, y_pred_binary)
    print(f"{name} - Accuracy: {accuracy:.2f}")
    print(classification_report(y_test_binary, y_pred_binary))
```

```python
# Feature importance using Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(x_train, y_train)
importances = rf_regressor.feature_importances_

# Plotting feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=x.columns)
plt.title("Feature Importance")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3, scoring='accuracy')
grid_search.fit(x_train, y_train_binary)

print(f"Best Parameters: {grid_search.best_params_}")
best_rf_model = grid_search.best_estimator_  # Retrieve the best model
```
